```python
def ols(xx,yy):
    # xx trials x vox
    return np.sum( (xx-np.mean(xx,0))*(yy-np.mean(yy,0)),0 )/np.sum( ((xx-np.mean(xx,0))**2),0 )
def std_e(x): return np.std(x,0)/np.sqrt(len(x))
def wrap(x):
    x[np.abs(x)>90]-=180*np.sign(x[np.abs(x)>90])
    return x

global obs, d_use_rad
def min_fun_SDT(p):
    VM = scipy.stats.vonmises(p[0],p[1])
    return -np.sum(np.log((1-obs)*VM.cdf(d_use_rad)+(obs)*(1-VM.cdf(d_use_rad))))


def do_bining(bns,overlap,grouping_var,var,var2=[-999],want_var=0):
    
    n_bns=len(bns)
    grouper = np.zeros(len(bns))
    if var2[0]!=-999:
        do_SDT=1
        out = np.zeros((2,len(bns)))

    else:
        do_SDT=0
        out = np.zeros(len(bns))
    for i in range(n_bns):
        if i<overlap:
            these_ind = (grouping_var<=bns[i+overlap]) | (grouping_var>=bns[i-overlap])
        elif i>(n_bns-overlap-1):
            these_ind = (grouping_var>=bns[i-overlap]) | (grouping_var<=bns[i+overlap-n_bns])
        else:
            these_ind = (grouping_var>=bns[i-overlap])&(grouping_var<=bns[i+overlap]) # need to figure out 
        if do_SDT:
            obs = var[these_ind].copy()
            d_use_rad = var2[these_ind].copy()
            fit = scipy.optimize.minimize(min_fun_SDT,[10,0],bounds=((0.001,50),(-pi/2,pi/2)))
            out[:,i] = [fit.x[0],-fit.x[1]/pi*90]
        else:
            if want_var==1:
                out[i] = np.var(var[these_ind])
            else:
                out[i] = np.mean(var[these_ind])
    return out
```

```python
def sem_plot(x,y,do_line=0,within_E=0,**args):
    # x is assumed to be examples x len(y)
    n_ex,n_points =  y.shape
    assert n_points==len(x), 'x not correct shape'
    m_y = np.mean(y,0)
    if within_E:
        y_use = (y.T-np.mean(y,1)).T
        s_y = np.std(y_use,0)/np.sqrt(n_ex)
    else:
        s_y = np.std(y,0)/np.sqrt(n_ex)
    plt.fill_between(x,m_y-s_y,m_y+s_y,**args)
    plt.plot(x,m_y,'k')
```

```python
global obs, d_use_rad

def min_fun_gauss(p):
    G = scipy.stats.norm(p[1],p[0])
    lik = (1-obs)*G.cdf(d_use_rad)+(obs)*(1-G.cdf(d_use_rad))
    return -np.sum(np.log(lik))

def do_bining(bns,overlap,grouping_var,var,var2=[-999],do_gauss=False,want_var=0,flip=0):
    global obs, d_use_rad
    n_bns=len(bns)
    grouper = np.zeros(len(bns))
    if var2[0]!=-999:
        do_SDT=1
        out = np.zeros((2,len(bns)))
        if flip:

            grouping_var = np.concatenate((grouping_var,-grouping_var))
            var = np.concatenate((var,(var==0)*1))
            var2 = np.concatenate((var2,-var2))
    else:
        do_SDT=0
        out = np.zeros(len(bns))
        if flip:
            grouping_var = np.concatenate((grouping_var,-grouping_var))
            if flip==2:
                var = np.concatenate((var,var))
            elif flip==1:
                var = np.concatenate((var,-var))
    for i in range(n_bns):
        if i<overlap:
            these_ind = (grouping_var<=bns[i+overlap]) | (grouping_var>=bns[i-overlap])
        elif i>(n_bns-overlap-1):
            these_ind = (grouping_var>=bns[i-overlap]) | (grouping_var<=bns[i+overlap-n_bns])
        else:
            these_ind = (grouping_var>=bns[i-overlap])&(grouping_var<=bns[i+overlap]) # need to figure out 
        if do_SDT:
            obs = var[these_ind].copy()
            d_use_rad = var2[these_ind].copy()
            if do_gauss:
                d_use_rad*=90/pi
                fit = scipy.optimize.minimize(min_fun_gauss,[12,0],bounds=((0.01,45),(-45,45)) )
                out[:,i] = [fit.x[0],-fit.x[1]]
            else:
                fit = scipy.optimize.minimize(min_fun_SDT,[10,0],bounds=((0.001,500),(-pi/2,pi/2)))
                out[:,i] = [fit.x[0],-fit.x[1]/pi*90]
                
        else:
            if want_var:
                out[i] = np.var(var[these_ind])
            else:
                out[i] = np.mean(var[these_ind])
    return out
```

```python
# get CC by timepoint & ROI
T=pd.DataFrame()
est_u = 'IEM' # IEM
for i,t in enumerate(t_use_try):
    for delay in [5,7,9,0]: # 5,7,9
        Sa_stim = SA[(SA.t_train==t)]
        if delay!=0:
            Sa = SA[(SA.t_train==t)&(SA.delay==delay)]
        else:
            Sa = Sa_stim
        Sl = SL

        for roi in roi_order:
            dec = Sa['est'+est_u+roi]
            stim = Sa['orient0']
#             decL = Sa['LestIEM'+roi]
#             decLL = Sl['estIEM'+roi]
            decL = Sa['Lest'+est_u+roi]
            decLL = Sl['est'+est_u+roi].values.copy()
            decLL[decLL<90]+=180
            decLL-=90
            stimL = Sl['d0']
            _cc= cc.circ_corr(stim+90,dec)
            _ccL= cc.circ_corr(stim+90,decL)
            _ccLL= cc.circ_corr(stimL+90,decLL)
            T = T.append({'subj':'Pool','roi':roi[:7],'delay':delay,'TR':t-6+2,'cc':_cc,'Lcc':_ccL,'LLcc':_ccLL,
                          'n':len(dec)},ignore_index=1)
            for subj in subjs:
                dec = Sa.loc[Sa.subj==subj,'est'+est_u+roi]
                decL = Sa.loc[Sa.subj==subj,'Lest'+est_u+roi]
#                 stim = Sa.loc[Sa.subj==subj,'orient0']
                if delay!=0:
                    stim_ind= Sa_stim[Sa_stim.subj==subj].delay.values==delay
                else:
                    stim_ind = np.ones(np.sum(Sa_stim.subj==subj))==1
                try:
                    stim = ori_task_all[subj][stim_ind]
                    decLL = Sl.loc[Sl.subj==subj,'estIEM'+roi].values.copy()
                    decLL[decLL<90]+=180
                    decLL-=90
                    stimL = Sl.loc[Sl.subj==subj,'d0']
                    _cc= cc.circ_corr(stim+90,dec)
                    _ccL= cc.circ_corr(stim+90,decL)
                    _ccLL= cc.circ_corr(stimL+90,decLL)
                    T = T.append({'subj':subj,'roi':roi[:7],'delay':delay,'TR':t-6+2,'cc':_cc,'Lcc':_ccL,'LLcc':_ccLL,'n':len(dec)},ignore_index=1)
                except:
                    print('Bad TP',subj,roi,t)
T['time'] = np.round((T['TR'])*.8,1)
```

# Figure S4A-B

```python
y_view = ['cc','Lcc'][0] #  flag, view memory (cc) or localizer (lcc) decoder

# do above but add sig testing
yl = [-.1,.25]

i,delay=0,0
t_view = 6.4 
t_view=0
xl = [-3.2,24]
ax1 = sns.lineplot(x='time',y=y_view,hue='roi',data=T[(np.isin(T.roi,roi_show))&(T.subj!='Pool')&(T.delay==delay)],ci=68)
ax1.legend_.remove()
plt.ylim(yl)
#     plt.xlim(xl)
plt.ylabel('$Circ_{corr}$')
plt.title('Loc N-1 CV (delay:%d)' %delay)
plt.plot([t_view,t_view],yl,'r')
plt.xlabel('')

plt.plot(xl,[0,0],'k-')
#     plt.plot(np.ones(2)*(delay),yl,'k--',alpha=.5)
#     if delay>0:
#         plt.plot(np.ones(2)*(avg_del[i-1]),yl,'k--',alpha=.5)
plt.xticks(xt)
plt.xlim(xl)


times_test= T.time.unique()
scores = np.zeros((len(rois),2,len(times_test)))
for ri,roi in enumerate(rois):
    for ti,t in enumerate(times_test):
        this_dat = T.loc[(T.roi==roi)&(T['time']==t)&(T.delay==0)&(T.subj!='Pool'),y_view].values
        this_dat = np.arctanh(this_dat) # fisher transform
#         this_dat = this_dat[np.abs(this_dat)<45]
        ts,ps = scipy.stats.ttest_1samp(this_dat,0)
        scores[ri,:,ti] = [ts,ps]

for ri in range(len(roi_show)):
    this_col = ax1.collections[ri].get_edgecolor()[0]
    this_col[-1] = .8
    this_score = scores[ri,1,:]/2 # one-tailed test
    this_score[(scores[ri,0,:]<0)] = 1.0 # only want + sig dec
        
    p05 = this_score<.05
    p01 = this_score<.01
    p001 = this_score<.001

    if y_view =='Lcc':
        this_y = ri*np.ones(len(p05))*.015 +.15#-.09
    else:
        this_y = ri*np.ones(len(p05))*.015 -.09
    this_y[~p05]=np.nan
    this_y_01 = this_y.copy()
    this_y_001 = this_y.copy()
    this_y_01[~p01] = np.nan
    this_y_001[~p001] = np.nan
    plt.plot(times_test,this_y,'.',color=this_col)
    plt.plot(times_test,this_y_01,'o',color=this_col,linewidth=4)
    plt.plot(times_test,this_y_001,'o',color=this_col,fillstyle='none',markersize=10)
    
    
# sav_fig('Dec_N-1_time_loc_stars')
```
