```python
def voxel_selection(samples,ev,params):
    # samples (voxels x trials)
    ind_all = np.arange(samples.shape[0])
    # donut selection
    donut_voxels = IEM.get_donut_mask(samples,ev,params['donut']) #minimun percentile
    donut_trials = (ev.d1.values==1)
    # anova selection
    samples_donut = samples[donut_voxels,:]
    samples_donut = samples_donut[:,donut_trials]
    ind_all_donut=ind_all[donut_voxels]
    
    ori = ev[donut_trials].d0.values
    ori_anova = np.round((ori-10)/2,-1)
    f_stats= IEM.anova1(samples_donut,ori_anova)
    n_keep = int(np.round(np.sum(donut_voxels)*params['anovaP']/100))
    ind_keep = np.argsort(f_stats)[-n_keep:]
    ind_all_both = ind_all_donut[ind_keep]
#     print('%d -> %d voxels' %(samples.shape[0],len(ind_all_both)))
    return ind_all_both, ind_all_donut



def gamma2(p):
    a1,loc1,amp1,a2,loc2,amp2 = p
    return amp1*gamma.pdf(t,a1,loc1)-amp2*gamma.pdf(t,a2,loc2)

def err_HRF(p,HRF):
    this_fit = gamma2(p)
    this_err = np.sum((HRF-this_fit)**2)
    return this_err

def wrap(x):
    x[np.abs(x)>90]-=180*np.sign(x[np.abs(x)>90])
    return x
def wrap_rad(x):
    x[np.abs(x)>pi]-=2*pi*np.sign(x[np.abs(x)>pi])
    return x
```

```python
k = np.sqrt(2)/np.exp(-0.5)
def DoG(p,x_min_deg):
    a,w = p
    y = x_min_deg*a*w*k*np.exp(-(w*x_min_deg)**2)
    return y

def min_fun_dog_neural(p,vals):
    x_min,eRecon = vals
    bias = DoG(p,x_min)
    return np.sqrt(np.sum(wrap(eRecon - DoG(p,x_min) )**2))
```

```python
def sem_plot(x,y,axs=0,do_line=0,within_E=0,do_sd=False,do_errorbars=False,**args):
    # x is assumed to be examples x len(y)
    n_ex,n_points =  y.shape
    n_ex = np.sum(~np.isnan(y),0)
    assert n_points==len(x), 'x not correct shape'
    m_y = np.nanmean(y,0)
    if within_E:
        y_use = (y.T-np.mean(y,1)).T
        s_y = np.nanstd(y_use,0)/np.sqrt(n_ex)
    else:
        s_y = np.nanstd(y,0)/np.sqrt(n_ex)
    if do_sd:
        s_y = np.std(y,0)
    if do_errorbars:
        plt.errorbar(x,m_y,s_y,**args)
        return
    if axs==0:
        plt.fill_between(x,m_y-s_y,m_y+s_y,**args)
        if do_line:
            plt.plot(x,m_y,**args)
    else:
        axs.fill_between(x,m_y-s_y,m_y+s_y,**args)
        axs.plot(x,m_y,**args)
```

```python
def do_bining(bns,overlap,grouping_var,var,flip=1,want_var=0,n_boot=1):

    n_bns=len(bns)
    
    out = np.zeros((len(bns),n_boot))
    if flip:
        grouping_var = np.concatenate((grouping_var,-grouping_var))
        if flip==3:
            var = np.concatenate((var,1-var))
        elif flip==2:
            var = np.concatenate((var,var))
        elif flip==1:
            var = np.concatenate((var,-var))
            
    for i in range(n_bns):
        
        if i<overlap:
            base_ind = (grouping_var<=bns[i+overlap]) | (grouping_var>=bns[i-overlap])
        elif i>(n_bns-overlap-1):
            base_ind = (grouping_var>=bns[i-overlap]) | (grouping_var<=bns[i+overlap-n_bns])
        else:
            base_ind = (grouping_var>=bns[i-overlap])&(grouping_var<=bns[i+overlap]) # need to figure out 
        base_ind_w = np.where(base_ind)[0]
        for b in range(n_boot):
            if n_boot==1:
                these_ind=base_ind
            else:
                these_ind = np.random.choice(base_ind_w,int(len(base_ind_w)*1),replace=True)
            
            
            if want_var==0:
                out[i] = np.mean(var[these_ind])
            elif want_var==1:
                out[i] = np.std(var[these_ind])
            elif want_var==2:
                out[i] = circ_sd(var[these_ind])
            elif want_var==2.5:
                out[i] = circ_var(var[these_ind])
            elif want_var==3:
                out[i] = circ_mean(var[these_ind])
    return out.squeeze()
```

```python
def select_voxel_set(HRF_FITS):
    # randomly select voxels to use out of non-problematic ones...
    nVox = HRF_FITS.shape[0]
    good_fits_stim = ~np.isnan(HRF_FITS[:,0,0])
    good_fits_probe= ~np.isnan(HRF_FITS[:,1,0])
    good_fits = good_fits_probe&good_fits_stim
    fits_use = np.where(good_fits)[0]
    activity_use = np.random.choice(fits_use,N_units,replace=True,) 
    return activity_use

def generate_stim_times(STICKS,N_stim):
    # get times that stim occur
    stims = np.hstack([s0 for s0,_ in STICKS])
    probes = np.hstack([s1 for _,s1 in STICKS])
    n_stim = np.sum(stims)
    #- will want to truncate expt to make consistent between subjects...
    if n_stim>N_stim:    # truncate! 
        last_tr = np.where(probes)[0][N_stim-1]+nTRModel+1
        stims = stims[:last_tr]
        probes = probes[:last_tr]
        stims[-nTRModel:]=0
        probes[-nTRModel:]=0
        
    n_stim = np.sum(stims)
    return (stims,probes)

def generate_response(STIMS,RESP,HRF_FIT,HRF_EMP,N_units,activity_use,adapt_rate=0.2,use_fit=0,noise_level=1.0):
    
    # gene
    stims,probes =STIMS
    n_stim,l_experiment = np.sum(stims), len(stims)
    stim_times,probe_times = np.nonzero(stims)[0],np.nonzero(probes)[0]
    
    stim_mat,probe_mat = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))
    stim_mat_adapt,probe_mat_adapt = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))
    
    # make stick reponses
    adapt_this = np.hstack((np.zeros((N_units,1)),resp_n[:,:-1]))*adapt_rate
    for i in range(N_units):
        stim_mat[i,stim_times] = RESP[0][i]
        probe_mat[i,probe_times] = RESP[1][i]

        stim_mat_adapt[i,stim_times] = RESP[0][i]-adapt_this[i]
        probe_mat_adapt[i,probe_times] = RESP[1][i]-adapt_this[i]
        
    # convolve with kernel
    stim_resp,probe_resp  = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))
    stim_resp_adapt,probe_resp_adapt  = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))

    l_kern = HRF_FIT.shape[2]
    shift = l_kern//2
    for i in range(N_units):
        if use_fit:
            this_kernel_stim = HRF_FIT[activity_use[i],0,:]
            this_kernel_probe = HRF_FIT[activity_use[i],1,:]
        else:
            this_kernel_stim = HRF_EMP[0][:,activity_use[i]]
            this_kernel_probe = HRF_EMP[1][:,activity_use[i]]

        stim_resp[i][shift:] = np.convolve(stim_mat[i],this_kernel_stim,mode='same')[:-shift]
        probe_resp[i][shift:] = np.convolve(probe_mat[i],this_kernel_probe,mode='same')[:-shift]
        stim_resp_adapt[i][shift:] = np.convolve(stim_mat_adapt[i],this_kernel_stim,mode='same')[:-shift]
        probe_resp_adapt[i][shift:] = np.convolve(probe_mat_adapt[i],this_kernel_probe,mode='same')[:-shift]
    
    response = stim_resp + probe_resp + noise_level*np.random.randn(*stim_resp.shape) # 
    response_adapt = stim_resp_adapt + probe_resp_adapt + noise_level*np.random.randn(*stim_resp.shape) # 
    return (response,response_adapt)


def decode_bias_x_time(sim_activity,d_ori,stim_times, tr_start = np.arange(0,20),epoch_len=4):
    response,response_adapt = sim_activity
    N_units = response.shape[0]
    n_tr_try = len(tr_start)
    bias_time = np.zeros((2,n_tr_try,n_bns))
    cc_time = np.zeros((2,n_tr_try))
    DoG_fits = np.zeros((2,n_tr_try,2))
    dog_bnds = ((-30,30),(.005,.1))
    for tri,tr in enumerate(tr_start):
        epoched_responses,epoched_responses_adapt = np.zeros((N_stim,N_units)),np.zeros((N_stim,N_units))
        for i in range(N_stim):
            this_tr = stim_times[i]+tr
            epoched_responses[i] = np.mean(response[:,this_tr:this_tr+epoch_len],1)
            epoched_responses_adapt[i] = np.mean(response_adapt[:,this_tr:this_tr+epoch_len],1)
        Y_hat = CRT.circ_regr_IEM_cval(epoched_responses[1:],Y,10)
        Y_hat_adapt = CRT.circ_regr_IEM_cval(epoched_responses_adapt[1:],Y,10)
        E = wrap(Y_hat-Y)
        E_adapt = wrap(Y_hat_adapt-Y)

        bias_time[0,tri] = do_bining(bns,overlap,d_ori,E,flip=0)
        bias_time[1,tri] = do_bining(bns,overlap,d_ori,E_adapt,flip=0)
        
        cc_time[0,tri] = CRT.circ_corr_coef(Y/90*pi,Y_hat/90*pi)
        cc_time[1,tri] = CRT.circ_corr_coef(Y/90*pi,Y_hat_adapt/90*pi)
    
        vals = (d_ori,E)
        this_fit = scipy.optimize.minimize(min_fun_dog_neural,(0,.03),args = (vals,),bounds=dog_bnds)
        DoG_fits[0,tri] = this_fit.x
        vals = (d_ori,E_adapt)
        this_fit = scipy.optimize.minimize(min_fun_dog_neural,(0,.03),args = (vals,),bounds=dog_bnds)
        DoG_fits[1,tri] = this_fit.x
    
    return (bias_time,cc_time,DoG_fits)

def deconvolve_activity(sim_activity,STIMS,HRF_FIT,HRF_EMP,activity_use,use_fit=0):
    response,response_adapt = sim_activity
    N_units,nTR_total = response.shape
    
    stim_times = np.nonzero(STIMS[0])[0]
    probe_times = np.nonzero(STIMS[1])[0]
    nTR=32

    # first fit for all voxels...
    coef_stim_all, coef_probe_all = np.zeros((N_units,N_stim)),np.zeros((N_units,N_stim))
    coef_stim_all_adapt, coef_probe_all_adapt= np.zeros((N_units,N_stim)),np.zeros((N_units,N_stim))
    for vox in range(N_units):
        if vox%10==0:
            print(vox,end = ' ')
        if use_fit:
            this_kernel_stim = HRF_FIT[activity_use[vox],0,:]
            this_kernel_probe = HRF_FIT[activity_use[vox],1,:]
        else:
            this_kernel_stim = HRF_EMP[0][:,activity_use[vox]]
            this_kernel_probe = HRF_EMP[1][:,activity_use[vox]]

        X_stim,X_probe = np.zeros((N_stim,nTR_total)),np.zeros((N_stim,nTR_total))

        for i in range(N_stim):
            nFit = np.min((nTR,nTR_total-stim_times[i]))
            X_stim[i,stim_times[i]:stim_times[i]+nTR] = this_kernel_stim[:nFit]
            X_probe[i,probe_times[i]:probe_times[i]+nTR] = this_kernel_probe[:nFit]
        X_Task = np.vstack((X_stim,X_probe))

        this_fit = LinearRegression().fit(X_Task.T,response[vox])
        fit_coef = this_fit.coef_ # 
        coef_stim_all[vox] = fit_coef[:N_stim]
        coef_probe_all[vox] = fit_coef[N_stim:]

        this_fit_adapt = LinearRegression().fit(X_Task.T,response_adapt[vox])
        fit_coef_adapt = this_fit_adapt.coef_ # 
        coef_stim_all_adapt[vox] = fit_coef_adapt[:N_stim]
        coef_probe_all_adapt[vox] = fit_coef_adapt[N_stim:]
    
    return ((coef_stim_all,coef_probe_all),(coef_stim_all_adapt,coef_probe_all_adapt))

def decode_all(DEC_deconvolve):
    FEATS_OG = DEC_deconvolve[0]
    FEATS_adapt = DEC_deconvolve[1]
    _DEC ,_E_ALL,_avg_bias = [],[],[]
    for FEAT in (FEATS_OG,FEATS_adapt):
        _DEC.append([CRT.circ_regr_IEM_cval(F.T[1:],Y,10) for F in FEAT])
        _E_ALL.append([wrap(y_hat-Y) for y_hat in _DEC[-1]])
        _avg_bias.append([do_bining(bns,overlap,d_ori,E) for E in _E_ALL[-1]])

    cc_all = np.zeros((2,2))
    bias_all = np.zeros((2,2,n_bns))

    for i in range(2):
        for want_adapt in range(2):
            cc_all[i,want_adapt] = CRT.circ_corr_coef(_DEC[want_adapt][i],Y)
            bias_all[i,want_adapt] = do_bining(bns,overlap  ,d_ori,_E_ALL[want_adapt][i],flip=0)
    return (bias_all,cc_all)
```

```python
# lets simulate task! 
si = 1
subj = subjs[si]
use_fit = 0 # 0- use raw deconv kernal. 1- use double gamma parameterization 
adapt_rate = 0.4 # percent adapt by
 
nVox=HRF_FITS_ALL[subj]['V2']['fits'].shape[0]
good_fits_stim = ~np.isnan(HRF_FITS_ALL[subj]['V2']['fits'][:,0,0])
good_fits_probe= ~np.isnan(HRF_FITS_ALL[subj]['V2']['fits'][:,1,0])
good_fits = good_fits_probe&good_fits_stim
fits_use = np.where(good_fits)[0]
activity_use = np.random.choice(fits_use,N_units,replace=True,)

STICKS = MDL_OUT_ALL[si][2]
stims = np.hstack([s0 for s0,_ in STICKS])
probes = np.hstack([s1 for _,s1 in STICKS])
n_stim = np.sum(stims)

# lets make stim_resp matrix
#- here assuming probe response will be of similar tuning and magnitude. 

#- will want to truncate expt to make consistent between subjects...
if n_stim>N_stim:    # truncate! 
    last_tr = np.where(probes)[0][N_stim-1]+nTRModel+1
    stims = stims[:last_tr]
    probes = probes[:last_tr]
    stims[-nTRModel:]=0
    probes[-nTRModel:]=0
    n_stim = np.sum(stims)
    
l_experiment = len(stims)
stim_times = np.where(stims)[0]
probe_times = np.where(probes)[0]
assert n_stim>=N_stim, 'Not enough trials!'

# response mag matrix
stim_mat,probe_mat = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))
stim_mat_adapt,probe_mat_adapt = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))

X_Task = np.zeros((2,l_experiment))
X_Task[0,stim_times]=1
X_Task[1,probe_times]=1

adapt_this = np.hstack((np.zeros((N_units,1)),resp_n[:,:-1]))*adapt_rate
for i in range(N_units):
    stim_mat[i,stim_times] = resp_n[i]
    probe_mat[i,probe_times] = resp_n_probe[i]
    
    stim_mat_adapt[i,stim_times] = resp_n[i]-adapt_this[i]
    probe_mat_adapt[i,probe_times] = resp_n_probe[i]-adapt_this[i]
    
#- convolve mat with HRFs

# now convolve with selected kernels, sum signals
stim_resp,probe_resp  = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))
stim_resp_adapt,probe_resp_adapt  = np.zeros((N_units,l_experiment)),np.zeros((N_units,l_experiment))

l_kern = HRF_FITS['V2']['fits'].shape[2]
shift = l_kern//2
for i in range(N_units):
    if use_fit:
        this_kernel_stim = HRF_FITS['V2']['fits'][activity_use[i],0,:]
        this_kernel_probe = HRF_FITS['V2']['fits'][activity_use[i],1,:]
    else:
        this_kernel_stim = MDL_OUT_ALL[si][0][roi][:,activity_use[i]]
        this_kernel_probe = MDL_OUT_ALL[si][1][roi][:,activity_use[i]]
        
    stim_resp[i][shift:] = np.convolve(stim_mat[i],this_kernel_stim,mode='same')[:-shift]
    probe_resp[i][shift:] = np.convolve(probe_mat[i],this_kernel_probe,mode='same')[:-shift]
    stim_resp_adapt[i][shift:] = np.convolve(stim_mat_adapt[i],this_kernel_stim,mode='same')[:-shift]
    probe_resp_adapt[i][shift:] = np.convolve(probe_mat_adapt[i],this_kernel_probe,mode='same')[:-shift]
```
