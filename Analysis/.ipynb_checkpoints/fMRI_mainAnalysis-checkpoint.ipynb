```python
import matplotlib 
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)

cmap = matplotlib.cm.get_cmap('Dark2')
ct = cmap(np.linspace(0,1,4))

def sav_fig(nam):
    root = './Figs/'
    plt.savefig(root + nam +'.svg',dpi=1200)
    
    
def circ_mean(x): return np.angle(np.mean(np.exp(1j*x/90*pi)))*90/pi # ang -> ang
def circ_mean2(x): return np.angle(np.mean(np.exp(1j*x/90*pi),1))*90/pi # ang -> ang

def circ_std(x):
    R = np.abs(np.mean(np.exp(1j*x/90*pi)))
    v = np.sqrt(-2*np.log(R))
    return v
def circ_var(x):
    R = np.abs(np.mean(np.exp(1j*x/90*pi)))
    v = 1-R
    return v
```

```python
# def ols(xx,yy):
#     # xx trials x vox
#     return np.sum( (xx-np.mean(xx,0))*(yy-np.mean(yy,0)),0 )/np.sum( ((xx-np.mean(xx,0))**2),0 )
def std_e(x): return np.std(x,0)/np.sqrt(len(x))
def wrap(x):
    x[np.abs(x)>90]-=180*np.sign(x[np.abs(x)>90])
    return x

global obs, d_use_rad
def min_fun_SDT(p):
    VM = scipy.stats.vonmises(p[0],p[1])
    return -np.sum(np.log((1-obs)*VM.cdf(d_use_rad)+(obs)*(1-VM.cdf(d_use_rad))))

def sem_plot(x,y,axs=0,do_line=0,within_E=0,**args):
    # x is assumed to be examples x len(y)
    n_ex,n_points =  y.shape
    assert n_points==len(x), 'x not correct shape'
    m_y = np.mean(y,0)
    if within_E:
        y_use = (y.T-np.mean(y,1)).T
        s_y = np.std(y_use,0)/np.sqrt(n_ex)
    else:
        s_y = np.std(y,0)/np.sqrt(n_ex)
    if axs==0:
        plt.fill_between(x,m_y-s_y,m_y+s_y,**args)
        if do_line:
            plt.plot(x,m_y,**args)
    else:
        axs.fill_between(x,m_y-s_y,m_y+s_y,**args)
        axs.plot(x,m_y,**args)
```

```python
GUESS_RATE = 0.25
def min_fun_gauss(p,dat_in):
    # sd, bias
    obs,d_use_deg = dat_in
    G = scipy.stats.norm(p[1],p[0])
    lik = ((1-obs)*G.cdf(d_use_deg)+(obs)*(1-G.cdf(d_use_deg)))*(1-GUESS_RATE)+0.5*GUESS_RATE
    return -np.sum(np.log(lik))

def do_bining(bns,overlap,grouping_var,var,var2=[-999],flip=1,prior_gauss=[4,0],want_var=0,n_boot=1):
    global obs, d_use_deg
    n_bns=len(bns)
    if var2[0]!=-999:
        do_SDT=1
        out = np.zeros((2,len(bns),n_boot))
            
        if flip:
            grouping_var = np.concatenate((grouping_var,-grouping_var))
            var = np.concatenate((var,(var==0)*1))
            var2 = np.concatenate((var2,-var2))
    else:
        do_SDT=0
        out = np.zeros((len(bns),n_boot))
        if flip:
            grouping_var = np.concatenate((grouping_var,-grouping_var))
            if flip==3:
                var = np.concatenate((var,1-var))
            elif flip==2:
                var = np.concatenate((var,var))
            elif flip==1:
                var = np.concatenate((var,-var))
            
    for i in range(n_bns):
        
        if i<overlap:
            base_ind = (grouping_var<=bns[i+overlap]) | (grouping_var>=bns[i-overlap])
        elif i>(n_bns-overlap-1):
            base_ind = (grouping_var>=bns[i-overlap]) | (grouping_var<=bns[i+overlap-n_bns])
        else:
            base_ind = (grouping_var>=bns[i-overlap])&(grouping_var<=bns[i+overlap]) # need to figure out 
        base_ind_w = np.where(base_ind)[0]
        for b in range(n_boot):
            if n_boot==1:
                these_ind=base_ind
            else:
                these_ind = np.random.choice(base_ind_w,int(len(base_ind_w)*1),replace=True)
            
            if do_SDT:
                dat_in = (var[these_ind],var2[these_ind])
                fit = scipy.optimize.minimize(min_fun_gauss,x0=prior_gauss,args=(dat_in,),bounds=((0.01,45),(-45,45)))
                out[:,i,b] = [fit.x[0],-fit.x[1]]
            else:
                if want_var==0:
                    out[i] = np.mean(var[these_ind])
                elif want_var==1:
                    out[i] = np.std(var[these_ind])
                elif want_var==2:
                    out[i] = circ_std(var[these_ind])
                elif want_var==2.5:
                    out[i] = circ_var(var[these_ind])
                elif want_var==3:
                    out[i] = circ_mean(var[these_ind])
    return out.squeeze()


def ori_plot(l=0,y=0):
    if l:
        plt.plot([-90,90],[0,0],'c-')
    plt.xlim(-90,90)
    plt.xticks([-90,-45,0,45,90])
    if y:
        plt.ylim(-90,90)
        plt.yticks([-90,-45,0,45,90])
```

```python
# fit DoG to behavior
k = np.sqrt(2)/np.exp(-0.5)

def DoG(p,x_min_deg):
    a,w = p
    y = x_min_deg*a*w*k*np.exp(-(w*x_min_deg)**2)
    return y

def min_fun_dog_neural(p,vals):
    x_min,eRecon = vals
    bias = DoG(p,x_min)
    return np.sqrt(np.sum(wrap(eRecon - DoG(p,x_min) )**2))

def min_fun_dog_percept(p,vals):
    resp, d_probe, x_min,fit_sd = vals
    crit = DoG(p,x_min)
    G = scipy.stats.norm(crit,fit_sd)
    p_cw = G.cdf(d_probe)
    lik = (p_cw*(resp==1)+(resp==0)*(1-p_cw))*(1-GUESS_RATE) + 0.5*GUESS_RATE
    return -np.sum(np.log(lik))


def fit_DoG_neural_subj(subj,roi,train='ERecon',time=-999,ind=-1,index_use=1): 
    if time==-999:
        if subj=='ALL': subj_ind = S.subj!=subj
        else: subj_ind = S.subj==subj
        this_S = S[good_task&subj_ind&index_use]
    else:
        if subj=='ALL': subj_ind = SA.subj!=subj
        else: subj_ind = SA.subj==subj
        this_S = SA[good_task_all&subj_ind&(SA.time==time)&index_use]
    
    d_this,good = get_n(this_S,ind)
    vals = (d_this[good],this_S[train+roi].values[good])

    fit_dog_this = scipy.optimize.minimize(min_fun_dog_neural,x0=(0,.03),args=((vals),))

    if (fit_dog_this.success==False)|(np.abs(fit_dog_this.x[0])>30)|(fit_dog_this.x[1]<.001)|(fit_dog_this.x[1]>.1):
        fit_dog_this = scipy.optimize.minimize(min_fun_dog_neural,x0=(0,.03),args=((vals),),method='Nelder-Mead')
        if (fit_dog_this.success==False)|(np.abs(fit_dog_this.x[0])>30)|(fit_dog_this.x[1]<.001)|(fit_dog_this.x[1]>.1):
            fit_dog_this = scipy.optimize.minimize(min_fun_dog_neural,x0=(0,.03),args=((vals),),bounds=((-30,30),(.001,.1)))
    return fit_dog_this.x

def fit_DoG_percept_subj(subj,ind=-1,index_use=1): 
    if subj=='ALL':
        these_ind = good_task & index_use
    else:
        these_ind = good_task&(S.subj==subj) & index_use
    this_S = S[these_ind]
    vals0 = (this_S['respCW'].values,this_S['d_stim'].values)
    
    #- first fit gaussian response model to get participant error width
    fit_gauss_all = scipy.optimize.minimize(min_fun_gauss,x0=(3,0),args=((vals0),),method='Nelder-Mead')
    
    #- then use this error distribution to estimate DoG bias
    if ind==-1:
        vals = (this_S['respCW'].values,-this_S['d_stim'].values,this_S['d_ori'].values,fit_gauss_all.x[0])
    else:
        d_this,good = get_n(this_S,ind)
        vals = (this_S['respCW'].values[good],-this_S['d_stim'].values[good],d_this[good],fit_gauss_all.x[0])
    fit_dog_this = scipy.optimize.minimize(min_fun_dog_percept,x0=(0,.03),args=(vals,),method='Nelder-Mead')

    return (fit_gauss_all.x,fit_dog_this.x)

def get_FWHM(fit_shape):
    n_samp,n_subj = fit_shape.shape
    hlf_amp = np.max(fit_shape,0)/2  
    d_hlf_min = np.abs(fit_shape-hlf_amp)
    hlf_inds = np.argsort(d_hlf_min,0)
    
    n_try = 10
    FWHM = np.zeros(n_subj)*np.nan
    for i in range(n_subj):
        these_inds = hlf_inds[:,i]
        ind0,ind1=these_inds[:2]
        for j in range(n_try):
            if np.abs(ind0-ind1)<10:
                ind1=these_inds[j+2]
            else:
                FWHM[i] = np.abs(bns_tight[ind0]-bns_tight[ind1])
    return FWHM


def get_n(S_in,ind): # just do for all S
    if ind<0:
        d_this = np.concatenate((np.zeros(-ind),wrap(S_in.orient0[:ind].values - S_in.orient0[-ind:].values)))
        good = (S_in.trial>1)&(~S_in.respCW.isna())#&(S_in.subj==subj)
    elif ind>0:
        d_this = np.concatenate((wrap(S_in.orient0[ind:].values - S_in.orient0[:-ind].values),np.zeros(ind)))
        good = (S_in.trial<=17-ind)&(~S_in.respCW.isna())#&(S_in.subj==subj)
    else:
        raise
    return (d_this,good)
```

```python
# get perceptual varianace for close versus far d_ori
df_thresh_match_all = pd.DataFrame()
n_reps = 121 # increase for more consistent stats...
thresh = 30
thresh_try = np.arange(20,41,5)
thresh_try = [30] # faster, comment out to try for different dOri
grp_name = ['Close','Far']
for subj in subjs:
    print(subj)
    these_ind = good_task.values&(S.subj.values==subj)
    for thresh in thresh_try:
        for grp in range(2): # inside, outside
            if grp==0:
                this_grp = np.abs(S['d_ori'].values)<thresh
                n0 = np.sum(these_ind&this_grp) # smaller group #
                n_lap = 1
            elif grp==1: 
                this_grp = np.abs(S['d_ori'].values)>thresh
                n_lap = n_reps
            ind_use = np.where(these_ind&this_grp)[0]
            n_this = np.sum(these_ind&this_grp)
            percent_correct = np.mean(S[these_ind&this_grp].correct)*100
            fits=np.zeros((n_lap,2))
            for i in range(n_lap):
                this_S = S.iloc[np.random.choice(ind_use,n0,replace=0)]

                # Flip values
                this_resp = this_S['respCW'].values.copy()
                this_d_stim = this_S['d_stim'].values.copy()
                this_resp[this_S.d_ori<0] = 1-this_resp[this_S.d_ori<0]
                this_d_stim[this_S.d_ori<0] *= -1

                vals0 = (this_resp,this_d_stim)
                fit_gauss_this = scipy.optimize.minimize(min_fun_gauss,x0=(3,0),args=((vals0),),method='Nelder-Mead')
                fits[i,:]= fit_gauss_this.x
            df_thresh_match_all = df_thresh_match_all.append({'thresh':thresh,'subj':subj,'d_ori':grp_name[grp],'precent':percent_correct,'sd':np.median(fits[:,0],0),'bias':np.median(fits[:,1],0)},ignore_index=1)
```

# verify bias f(dTheta) effects are not driven by cardinal effects

```python
card_thresh = 22.5
card_ori = (np.abs(SA.orient0.values)<card_thresh) | (np.abs(wrap(SA.orient0.values-90))<card_thresh)
close_prev = np.abs(SA.d_ori.values)<30

prop_card = np.zeros((len(subjs),2))
for si,subj in enumerate(subjs):
    these_ind = good_all.values&(SA.subj.values==subj)
    prop_card[si,0] = np.sum((these_ind&close_prev&card_ori))/np.sum((these_ind&close_prev))
    prop_card[si,1] = np.sum((these_ind&~close_prev&card_ori))/np.sum((these_ind&~close_prev))
```
